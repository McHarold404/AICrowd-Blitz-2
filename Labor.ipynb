```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
```

```python
cross_validate=train_data.loc[25000:31997]
train_data=train_data.loc[0:25001]
train_data
```

```python
train_data.isnull().sum()
```

```python
y_train=train_data['class']
train_data=train_data.drop(columns=['class'])
cross_validate_y=cross_validate['class']
cross_validate=cross_validate.drop(columns=['class'])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(train_data,y_train,test_size=0.2,random_state=42)
#sc=StandardScaler()
#X_train=sc.fit_transform(X_train)
#X_test=sc.transform(X_test)
```

```python
#Random Classifier.
#from sklearn.datasets import make_classification
#X_train,Y_train=make_classification()
rfc=RandomForestClassifier()
#rfc=svm.SVC()
rfc.fit(X_train,Y_train)
pred_rfc=rfc.predict(X_test)
pred_rfc2=rfc.predict(cross_validate)
```

```python
#print(classification_report(Y_test,pred_rfc))
print(classification_report(cross_validate_y,pred_rfc2))
```

```python
submission=rfc.predict(test_data)
submission=pd.DataFrame(submission)
submission.to_csv('submission.csv',header=['class'],index=False)
```
